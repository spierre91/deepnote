```
import pandas as pd
import numpy as np 

#Define Product Name 
products = ["iPhone XR", "Galaxy S10+", "LG G8 ThinQ", "Moto G7 Power", "Pixel 3A XL",
            "Huawei P30 Pro", "Lenovo K8 Note", "Nokia 8.1", "Xiaomi Pocophone F1", "Oppo F11"
            ] 

#Define product codes
product_codes = [9288935, 9475345, 9337623, 9225232, 9403029,
                 9313, 6522, 9188955, 9575769, 9254999, 9534777]

#Define product descriptions 
product_desc = ["High-Performance Smartphone", "On-the-Go Flagship Phone", "Powerful Device",
                "Highly Portable Phone", "Camera-Focused Phone", "High-End Photography", 
                "Designed for Gamers", "All-in-One Package", "Next-Gen Specs", "Best-Value Option"
                ] 

#Set date range 
start = pd.to_datetime('2010-01-01')
stop = pd.to_datetime('2023-12-31')

#Generate Datetime values
dates = np.random.choice(pd.date_range(start, stop), 10000)

# Generate units_sold values 
units_sold = np.random.normal(loc=1000, scale=200, size=10000)
int_units = np.round(units_sold)

# Generate price values with negative linear correlation to units sold 
price = int_units * -0.5 + np.random.normal(loc=500, scale=80, size=10000)

# Generate discount values with positive linear correlation to units sold 
discount = int_units * 0.2 + np.random.normal(loc=100, scale=30, size=10000)

# Generate promotion values 
promotion = np.random.normal(loc=10, scale=2, size=10000)

# Generate brand loyalty values 
brand_loyalty = np.random.lognormal(mean=1.08, sigma=0.2, size=10000)

#Generate advertising campaigns values 
advertising = np.random.gamma(shape=2, scale=10, size=10000)

#Generate availability of products value
availability = np.random.negative_binomial(n=2, p=0.4, size=10000)

#Generate population density of area 
pop_density = np.random.beta(a=2, b=5, size=10000)

#Generate market trends
market_trends = np.random.gamma(shape=1.5, scale=2, size=10000)

#Generate weather in the area
weather = np.random.poisson(lam=3, size=10000)

#Generate local competitors in the area 
local_competitors = np.random.triangular(left=1, mode=20, right=50, size=10000)

#Generate socioeconomic status
socio_status = np.random.normal(loc=3, scale=1.5, size=10000)

#Generate geography 
geography = np.random.triangular(left=0, mode=2, right=4, size=10000)

#Generate customer demographics
customer_demo = np.random.logistic(loc=0, scale=1.2, size=10000)

#Generate mobile app download/usage
mobile_usage = np.random.randint(low=1, high=20, size=10000)

#Create Dataframe
data = {'Product': np.random.choice(products, 10000), 
        'product_codes': np.random.choice(product_codes, 10000), 
        'product_desc': np.random.choice(product_desc, 10000), 
        'date_time': dates, 
        'units_sold': int_units, 
        'price': price, 
        'discount': discount, 
        'promotion': promotion,
        'brand_loyalty': brand_loyalty, 
        'advertising': advertising, 
        'availability': availability, 
        'pop_density': pop_density, 
        'market_trends': market_trends, 
        'weather': weather, 
        'local_competitors': local_competitors, 
        'socio_status': socio_status, 
        'geography': geography, 
        'customer_demo': customer_demo, 
        'mobile_usage': mobile_usage
        }

df = pd.DataFrame(data)
df.head()
```

```
from catboost import CatBoostRegressor
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
# separate X and y
features = ['WEEK', 'MONTH', 'YEAR', 'Product', 'product_codes', 'product_desc', 
       'price', 'discount', 'promotion', 'brand_loyalty', 'advertising',
       'availability', 'pop_density', 'market_trends', 'weather',
       'local_competitors', 'socio_status', 'geography', 'customer_demo',
       'mobile_usage']

best_buy_df['date_time'] = pd.to_datetime(best_buy_df['date_time'])
best_buy_df['WEEK'] = best_buy_df['date_time'].dt.week
best_buy_df['MONTH'] = best_buy_df['date_time'].dt.month
best_buy_df['YEAR'] = best_buy_df['date_time'].dt.year

df_train = best_buy_df[best_buy_df['YEAR'] < 2018]
df_test = best_buy_df[best_buy_df['YEAR'] > 2018]


X_train = df_train[features]
y_train = df_train['units_sold']

X_test = df_test[features]
y_test = df_test['units_sold']

cats = ['Product', 'product_desc',]


# initialize catboost regressor 
cat_reg = CatBoostRegressor(iterations=1000,
                            cat_features = cats,
                            learning_rate=0.3,
                            depth=8,
                            eval_metric='MAPE',
                            random_seed=42)

# fit the model 
cat_reg.fit(X_train, y_train)

# predict values
y_pred = cat_reg.predict(X_test)
```
